```python
import pandas as pd
import plotly.express as px 
from itables import show
import numpy as np
#import libraries

cancer_data = pd.read_csv("data/breast_cancer_number_of_new_female_cases.csv") #load data

cancer_long = cancer_data.melt(id_vars=['geo', 'name'], var_name='year', value_name='cases')

cancer_long = cancer_long.rename(columns={'geo': 'country_code', 'name': 'country'}) #rename column heads

cancer_long['year'] = pd.to_numeric(cancer_long['year'])

latest_year = cancer_long['year'].max()
earliest_year = cancer_long['year'].min()


data_2023 = cancer_long[cancer_long['year'] == latest_year]
data_1990 = cancer_long[cancer_long['year'] == earliest_year] #filter for lastest year and first year

# Calculate values for cards
total_cases_2023 = data_2023['cases'].sum()
avg_cases_2023 = data_2023['cases'].mean()
max_cases_val = data_2023['cases'].max()
max_country = data_2023.loc[data_2023['cases'].idxmax(), 'country']
latest_stat = cancer_long[cancer_long['year'] == latest_year]
top_country_row = latest_stat.sort_values(by='cases', ascending=False).iloc[0]
top_country = top_country_row['country']
top_country_value = top_country_row['cases']
trend = cancer_long.groupby('year')['cases'].sum().reset_index()

area = px.area(trend, x='year', y='cases', title="Total female breast cancer cases globally")

top_10 = data_2023.nlargest(10, 'cases')

bar1 = px.bar(top_10, x='cases', y='country', orientation='h', color='cases', color_continuous_scale='Reds',title="Top 10 impacted countries in 2023")

map = px.choropleth(data_2023, locations="country_code", color="cases", hover_name="country",color_continuous_scale='Viridis', title="Global heatmap of breast cancer cases (2023)")

bar2 = px.bar(data_2023.sample(10), x='country', y='cases', color='country', title="Sample comparison of 10 countries")
```
